### Data Reading and Initial Visualization

#### IHDP Dataset description:

The Infant Health Development Program (IHDP) dataset was checked to see the effect of high quality childcare and home visits on the future cognitive test score of low-birth-weight, pre mature infants. The dataset contains 25 features, including measurements about the child (e.g., birth,weight, head diameter , weeks born,birth order,first born,neonatal health index, sex...) and mother info when she gave birth (e.g., age, marital status, education,work) and her behaviour (e.g., Smoke,Drink,Druges...). VAriable X are these. The treatment variable (t) indicates whether a family was part of the control (i.e., t = 0, no support was provided) or the treatment (i.e., t = 1, support was provided) group. The outcome column shows the cognitive score for the child. We use a semi-synthetic both factual and counterfactual outcomes version of the data are simulated based on real pre-treatment covariates. For this reason, the dataset also includes true individualised effects per each data unit, which are better suited for performance evaluation than the outcomes due to lack of noise. The use of counterfactuals/true effects is forbidden in the training stage.

#### JOBS Dataset description:

This dataset, proposed by [3], is a combination of the experiment done by [4] as part of the National Supported Work Program (NSWP) and observational data from the Panel Study of Income Dynamics (PSID) [5]. Overall, the data captures people’s basic characteristics via 17 variables about their background, whether they received job training from NSWP (treatment), and their employment status (outcome). Information about whether a sample comes from experimental or observational data is recorded under column ‘e’, though all records should be used for modelling.

#Decision making

By looking at both datasets, we can conclude that there is not too much pre-processing to do, because there are no null-values and no categorical features. However, we can see that IHDP dataset has factuals and counterfactuals (RCTs), and the JOBS dataset only has factuals (Mixture of experimental and observational data). With this information, for the IHDP dataset we can use the following metrics: -ATE -PEHE

And for JOBS dataset we must use the following metrics: -ATT -Policy Risk

Metric functions to evaluate IHDP dataset (ATE and PEHE)

```python
#lab 4 code of the module to evaluate with both metrics.
def abs_ate(effect_true, effect_pred):
    return np.abs(mean(effect_pred) - np.mean(effect_true))

def pehe(effect_true, effect_pred):
    return np.sqrt(np.mean((effect_true-effect_pred)**2))
```

## Metric functions to evaluate JOBS dataset (ATT and Policy Risk)

```python
#These functions were extracted from https://github.com/dmachlanski/CE888_2022/blob/main/project/metrics.py
def abs_att(effect_pred, yf, t, e):
    """
    Absolute error for the Average Treatment Effect on the Treated
    :param effect_pred: predicted treatment effect value
    :param yf: factual (observed) outcome
    :param t: treatment status (treated/control)
    :param e: whether belongs to the experimental group
    :return: absolute error on ATT
    """
    att_true = np.mean(yf[t > 0]) - np.mean(yf[(1 - t + e) > 1])
    att_pred = np.mean(effect_pred[(t + e) > 1])

    return np.abs(att_pred - att_true)
def policy_risk(effect_pred, yf, t, e):
    """
    Computes the risk of the policy defined by predicted effect
    :param effect_pred: predicted treatment effect value
    :param yf: factual (observed) outcome
    :param t: treatment status (treated/control)
    :param e: whether belongs to the experimental group
    :return: policy risk
    """
    # Consider only the cases for which we have experimental data (i.e., e > 0)
    t_e = t[e > 0]
    yf_e = yf[e > 0]
    effect_pred_e = effect_pred[e > 0]

    if np.any(np.isnan(effect_pred_e)):
        return np.nan

    policy = effect_pred_e > 0.0
    treat_overlap = (policy == t_e) * (t_e > 0)
    control_overlap = (policy == t_e) * (t_e < 1)

    if np.sum(treat_overlap) == 0:
        treat_value = 0
    else:
        treat_value = np.mean(yf_e[treat_overlap])

    if np.sum(control_overlap) == 0:
        control_value = 0
    else:
        control_value = np.mean(yf_e[control_overlap])

    pit = np.mean(policy)
    policy_value = pit * treat_value + (1.0 - pit) * control_value

    return 1.0 - policy_value
```
